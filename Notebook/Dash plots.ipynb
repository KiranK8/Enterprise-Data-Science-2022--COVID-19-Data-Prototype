```python
import pandas as pd
import numpy as np
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_test=pd.read_csv('Documents/ads_covid-19/data/processed/Covid_full_data.csv',sep=';')
df_test.head()
```

```python
country_list=['Italy',
              'India',
              'Spain',
              'Germany',
              'United States',
             ] 
```

```python
df_test2  = pd.DataFrame()
```

```python
for each in country_list:
    df_test2= df_test2.append(df_test[df_test['location']==each])
```

```python
df_test2=df_test2.reset_index(drop=True)   
```

```python
df_test2
```

```python

sel_countries = df_test2[df_test2['location']=='India']
sel_countries.head()
```

# The relative cases overtime of Covid infectors for 5 Countries

```python
import dash
dash.__version__
from dash import dcc
from dash import html

app = dash.Dash()
app.layout = html.Div([
    
    html.Label('Multi-Select Country'),
    
       dcc.Dropdown(
        id='country_drop_down',
        options=[{'label': i, 'value': i}
                for i in df_test2['location'].unique() ],
           
        value=['India','Germany'], # which are pre-selected
        multi=True
    ),   
        
    dcc.Graph(id='main_window_slope', figure={})
])
```

```python
from dash.dependencies import Input, Output

@app.callback(
    Output('main_window_slope', 'figure'),
    [Input('country_drop_down', 'value')])
def update_graph(option):
    print(option)
    print(type(option))
    
    if type(option)!=str:
        dff = df_test2[df_test2['location'].isin(option) ]
    else:
        dff = df_test2[df_test2['location']==option]
        

    # dff = df_test2[df_test2["location"] == option]

    # Plotly Express
    
    fig = px.line(dff, x='date', y='Relative_cases',color='location',title='Covid Relative Cases Over Time')
    
    return fig
```

```python
app.run_server(debug=True, use_reloader=False)
```
